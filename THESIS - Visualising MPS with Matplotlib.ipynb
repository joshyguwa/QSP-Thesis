```python
import matplotlib.pyplot as plt

# Enable LaTeX rendering for text
plt.rcParams['text.usetex'] = True

# Number of tensors in the chain (sites)
num_tensors = 7

# Positions for the tensors (nodes)
x_pos = list(range(num_tensors))
y_pos = [0] * num_tensors

# Define the bipartition index (between tensor index 3 and 4, i.e., A4 and A5)
bipartition = 4  # 0-based indexing

# Create a wider figure and axis
fig, ax = plt.subplots(figsize=(12, 6))  # Adjust the figsize as needed

# Plot tensors as circles
ax.scatter(x_pos, y_pos, s=1400, c='#FCE1A4', edgecolors='black', zorder=2)

# Add labels for tensors
for i, (x, y) in enumerate(zip(x_pos, y_pos)):
    ax.text(x, y, rf'$A^{{[{i+1}]}}$', ha='center', va='center', fontsize=18)

# Plot lines (bonds) between tensors and add text along the lines
for i in range(num_tensors - 1):
    # Plot the line (bond) between tensors
    ax.plot([x_pos[i], x_pos[i+1]], [y_pos[i], y_pos[i+1]], c='black', zorder=1)
    
    # Calculate the midpoint for adding text
    midpoint_x = (x_pos[i] + x_pos[i+1]) / 2
    midpoint_y = (y_pos[i] + y_pos[i+1]) / 2
    
    # Add LaTeX-style text for each bond
    ax.text(midpoint_x, midpoint_y + 0.05, rf'$\alpha_{{{i+1}}}$', ha='center', va='center', fontsize=16, color='black')

# ------------------ SVD Visualization ------------------

# Positions for SVD tensors
svd_x = [x_pos[bipartition - 1] + 0.5] * 3  # Placed between A4 and A5
svd_y = [0.5, 0, -0.5]  # U above, Σ in the middle, V† below

# Labels for SVD tensors
svd_labels = [r'$U$', r'$\Sigma$', r'$V^\dagger$']

# Plot SVD tensors as circles
ax.scatter(svd_x, svd_y, s=1200, c='#A4D4FCE1', edgecolors='black', zorder=3)

# Add labels for SVD tensors
for i, (x, y) in enumerate(zip(svd_x, svd_y)):
    ax.text(x, y, svd_labels[i], ha='center', va='center', fontsize=16)

# Connect A4 to U, U to Σ, Σ to V†, and V† to A5
# A4 to U
ax.plot([x_pos[bipartition -1], svd_x[0]], [y_pos[bipartition -1], svd_y[0]], c='black', zorder=1)
# U to Σ
ax.plot([svd_x[0], svd_x[1]], [svd_y[0], svd_y[1]], c='black', zorder=1)
# Σ to V†
ax.plot([svd_x[1], svd_x[2]], [svd_y[1], svd_y[2]], c='black', zorder=1)
# V† to A5
ax.plot([svd_x[2], x_pos[bipartition]], [svd_y[2], y_pos[bipartition]], c='black', zorder=1)

# Optionally, highlight the bipartition bond
ax.plot([x_pos[bipartition -1], x_pos[bipartition]], [y_pos[bipartition -1], y_pos[bipartition]], 
        c='red', linewidth=2, linestyle='--', zorder=0)

# Add a label for the bipartition
ax.text((x_pos[bipartition -1] + x_pos[bipartition]) / 2, 0.3, 
        'Bipartition and SVD', ha='center', va='bottom', fontsize=14, color='red')

# Set axis limits and turn off the axes
ax.set_xlim(-0.5, num_tensors - 0.5)
ax.set_ylim(-1.5, 1.5)
ax.axis('off')

# Add a title
ax.set_title('Matrix Product State (MPS) with SVD Across Bipartition', fontsize=16)

# Save the figure
plt.savefig('MPS_SVD_Visualization.png', dpi=600, transparent=True, bbox_inches='tight')

# Display the plot
plt.show()
```
