```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from matplotlib import rcParams

# Enable LaTeX-style rendering
rcParams.update({
    "text.usetex": True,
    "font.family": "serif",
    "font.serif": ["Computer Modern"],
})

# Create a new figure and axis with equal aspect ratio
fig, ax = plt.subplots(figsize=(6, 6))
ax.set_aspect('equal')

# Define properties for the larger circle (Hilbert Space)
large_circle = patches.Circle(
    (0.5, 0.5), 0.25,
    edgecolor='#1f77b4',      # A pleasant blue
    facecolor='#aec7e8',      # Light blue fill
    linewidth=2,
    alpha=0.8,                # Slight transparency
    zorder=1                  # Ensure it's below the ellipse
)

# Define properties for the smaller ellipse (Achievable States)
small_ellipse = patches.Ellipse(
    (0.5, 0.35), 0.25, 0.15,  # (center_x, center_y), width, height
    edgecolor='#ff7f0e',        # A warm orange
    facecolor='#ffbb78',        # Light orange fill
    linewidth=2,
    alpha=0.9,                  # Slight transparency
    zorder=2                    # Ensure it's above the large circle
)

# Add shapes to the axis
ax.add_patch(large_circle)
ax.add_patch(small_ellipse)

# Add labels using LaTeX formatting
ax.text(
    0.5, 0.5,
    r'\textbf{Hilbert Space} $\mathcal{H}$',
    horizontalalignment='center',
    verticalalignment='center',
    fontsize=30,
    color='black',
    zorder=3
)

ax.text(
    0.5, 0.35,
    r'\textbf{Achievable States}',
    horizontalalignment='center',
    verticalalignment='center',
    fontsize=22,
    color='black',
    zorder=4
)

# Set limits to ensure all elements are visible
ax.set_xlim(0.24, 0.76)  # Adjusted to tightly fit the large circle
ax.set_ylim(0.24, 0.76)  # Adjusted to tightly fit the large circle

# Remove axes for a cleaner look
ax.axis('off')

# Optional: Add a subtle background grid or gradient (enhances aesthetics)
# For simplicity, we'll skip this step, but it can be added if desired.

# Display the plot
plt.tight_layout()

plt.savefig(r'C:\Users\0josh\OneDrive\Desktop\Re-organised Honours Year\Thesis Documents\Misc. Plots\dimensionality_burden.png', dpi=600, bbox_inches='tight', transparent=True)

plt.show()
```
