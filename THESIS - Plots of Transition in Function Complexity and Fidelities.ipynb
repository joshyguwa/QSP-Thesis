```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import FancyArrowPatch
from scipy.ndimage import gaussian_filter1d
from scipy.interpolate import make_interp_spline

# Configure Matplotlib to use LaTeX-style fonts
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "serif",
    "font.serif": ["Computer Modern"],
    "axes.labelsize": 18,
    "xtick.labelsize": 0,  # Remove x-axis tick labels
    "ytick.labelsize": 0,  # Remove y-axis tick labels
})

# Set parameters
n = 8
num_points = 2**n
x = np.linspace(0, 4 * np.pi, num_points)

# Plot 1: Smooth, Well-Behaved Function (Sine Wave)
y1 = np.sin(x)

# Plot 2: Smooth Random Fluctuations Between 0 and 12
np.random.seed(0)  # For reproducibility
random_noise = np.random.normal(0, 1, num_points)  # Generate random noise

# Apply Gaussian filter to smooth the noise
sigma = 2  # Adjusted for smoother noise
smoothed_noise = gaussian_filter1d(random_noise, sigma=sigma)

# Normalize the smoothed noise to range between 0 and 12
y2_min, y2_max = smoothed_noise.min(), smoothed_noise.max()
y2 = (smoothed_noise - y2_min) / (y2_max - y2_min) * 12

# Plot 3: Smooth, Differentiable Curve Fit to Random Points
np.random.seed(0)  # For reproducibility
y_random = np.random.uniform(-2, 2, num_points)

# Create a cubic spline that interpolates the random points
spline = make_interp_spline(x, y_random, k=3)
y3 = spline(x)

# Create subplots
fig, axs = plt.subplots(1, 3, figsize=(21, 6))

# Plot 1: Smooth Sine Wave
axs[0].plot(x, y1, color='#d1db9c')
axs[0].set_title('Highly Structured Vector',fontsize=22)  # Added Title
axs[0].set_xlabel('$x$')
axs[0].set_ylabel('$f(x)$')
axs[0].set_xticks([])  # Remove x-axis ticks
axs[0].set_yticks([])  # Remove y-axis ticks

# Plot 2: Smooth Random Fluctuations Between 0 and 12
axs[1].plot(x, y2, color='#7fcdbb')
axs[1].set_title('Transition Period',fontsize=22)  # Added Title
axs[1].set_xlabel('$x$',weight='bold')
axs[1].set_ylabel('$f(x)$',weight='bold')
axs[1].set_ylim([0, 12])
axs[1].set_xticks([])  # Remove x-axis ticks
axs[1].set_yticks([])  # Remove y-axis ticks

# Plot 3: Smooth Spline Fit to Random Points
axs[2].plot(x, y3, color='#2c7fb8')
axs[2].set_title('Unstructured Vector',fontsize=22)  # Added Title
axs[2].set_xlabel('$x$',weight='bold')
axs[2].set_ylabel('$f(x)$',weight='bold')
axs[2].set_xticks([])  # Remove x-axis ticks
axs[2].set_yticks([])  # Remove y-axis ticks

# Adjust layout for better spacing
plt.tight_layout()

# Increase horizontal space to make room for arrows
fig.subplots_adjust(wspace=0.3)

# Function to add an arrow between two axes with padding
def add_arrow(ax1, ax2, fig, text=None, text_pos=(0, 0.05), delta=0.01):
    """
    Adds an arrow from ax1 to ax2 on the figure with specified padding.

    Parameters:
    - ax1: Starting Axes object
    - ax2: Ending Axes object
    - fig: Figure object
    - text: Optional text label for the arrow
    - text_pos: Position adjustment for the text label
    - delta: Padding offset to create white space around the arrow
    """
    # Get the bounding boxes of the axes in figure coordinates
    bbox1 = ax1.get_position()
    bbox2 = ax2.get_position()

    # Calculate start and end points for the arrow with padding
    start_x = bbox1.x1 - delta  # Slightly inside the first subplot
    start_y = bbox1.y0 + bbox1.height / 2
    end_x = bbox2.x0 + delta     # Slightly inside the second subplot
    end_y = bbox2.y0 + bbox2.height / 2

    # Create the arrow
    arrow = FancyArrowPatch(
        (start_x, start_y),
        (end_x, end_y),
        transform=fig.transFigure,
        arrowstyle="->",
        color='black',
        mutation_scale=20,
        linewidth=4
    )
    fig.patches.append(arrow)

    # Add text if provided
    if text:
        fig.text(
            (start_x + end_x) / 2 + text_pos[0]*0.02,
            (start_y + end_y) / 2 + text_pos[1]*0.02,
            text,
            ha='center',
            va='center',
            fontsize=12,
            backgroundcolor='white'
        )

# Add arrows between Plot 1 to Plot 2 and Plot 2 to Plot 3 with padding
add_arrow(axs[0], axs[1], fig, text_pos=(0, 0.05), delta=-0.02)
add_arrow(axs[1], axs[2], fig, text_pos=(0, 0.05), delta=-0.02)

plt.savefig(r'C:\Users\0josh\OneDrive\Desktop\Re-organised Honours Year\Thesis Documents\Misc. Plots\Entropy_Transition.png', dpi=300, bbox_inches='tight',transparent=True)

# Display the plots
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad
from matplotlib.patches import FancyArrowPatch

# Enable LaTeX-style text rendering in matplotlib
plt.rcParams['text.usetex'] = True
plt.rcParams['font.family'] = 'serif'

# 1. Define the Gaussian Function
def gaussian(x, mu=0, sigma=1):
    return (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-((x - mu) ** 2) / (2 * sigma ** 2))

# Parameters
mu = 0              # Mean
sigma = 1           # Standard deviation
x_min, x_max = -3, 3  # Focused range
x = np.linspace(x_min, x_max, 1000)
y = gaussian(x, mu, sigma)

# Function to calculate bin heights as function values at bin centers
def calculate_bin_heights(bin_edges):
    bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2
    y_bin_centers = gaussian(bin_centers, mu, sigma)
    bin_heights = y_bin_centers  # Alternatively, use probability mass
    return bin_centers, bin_heights, y_bin_centers

# Function to calculate bin heights as probability mass within each bin
def calculate_bin_areas(bin_edges):
    bin_areas = np.array([quad(gaussian, bin_edges[i], bin_edges[i+1])[0] for i in range(len(bin_edges)-1)])
    return bin_areas

# 2. Discretize the Gaussian into 8 bins (3-qubit representation) within [-3, 3]
qubits_3 = 3
bins_3 = 2 ** qubits_3  # 8 bins
bin_edges_3 = np.linspace(x_min, x_max, bins_3 + 1)
bin_centers_3, bin_heights_3, y_bin_centers_3 = calculate_bin_heights(bin_edges_3)
# Optionally, use probability mass instead of function values
# bin_heights_3 = calculate_bin_areas(bin_edges_3)

# 3. Discretize the Gaussian into 16 bins (4-qubit representation) within [-3, 3]
qubits_4 = 4
bins_4 = 2 ** qubits_4  # 16 bins
bin_edges_4 = np.linspace(x_min, x_max, bins_4 + 1)
bin_centers_4, bin_heights_4, y_bin_centers_4 = calculate_bin_heights(bin_edges_4)
# Optionally, use probability mass instead of function values
# bin_heights_4 = calculate_bin_areas(bin_edges_4)

# 4. Plot the Gaussian and both Discretizations Side by Side
# Increased figure size for better spacing
fig, axes = plt.subplots(1, 2, figsize=(18, 8), sharey=True)

# Common Gaussian curve for both plots
for ax in axes:
    ax.plot(x, y, label=r'$n = 4$', color='#003147', linewidth=2)

# --- Left Plot: 3-Qubit (8-Bin) Discretization ---
ax = axes[0]
# Plot discretized bar graph
bar_width_3 = (bin_edges_3[1] - bin_edges_3[0]) * 1  # Slightly narrower for better visibility
ax.bar(bin_centers_3, bin_heights_3, width=bar_width_3,
       alpha=0.6, color='#FED976', edgecolor='black', label=r'$n = 3$')

# Add black dots at each discretization point
ax.scatter(bin_centers_3, y_bin_centers_3, color='black', zorder=5, label='Discretization Points', s=50)

# Labels and Title for Left Plot
ax.set_xlabel(r'$x$', fontsize=28)  # Increased axis label font size
ax.set_ylabel(r'$p(x)$', fontsize=28)  # Added y-axis label with increased font size
ax.set_title(r'$n = 3$', fontsize=30)  # Increased title font size
#ax.legend(fontsize=16)  # Increased legend font size
ax.grid(False)
ax.set_xlim(x_min, x_max)
ax.set_ylim(0, max(y_bin_centers_4)*1.2)  # Adjust y-limit based on 4-qubit for consistency

# Set tick label sizes
ax.tick_params(axis='both', labelsize=18)

# --- Right Plot: 4-Qubit (16-Bin) Discretization ---
ax = axes[1]
# Plot discretized bar graph
bar_width_4 = (bin_edges_4[1] - bin_edges_4[0]) * 1  # Slightly narrower for better visibility
ax.bar(bin_centers_4, bin_heights_4, width=bar_width_4,
       alpha=0.6, color='#FEB24C', edgecolor='black', label=r'$n = 4$')

# Add red dots at each discretization point
ax.scatter(bin_centers_4, y_bin_centers_4, color='black', zorder=5, label='Discretization Points', s=30)

# Labels and Title for Right Plot
ax.set_xlabel(r'$x$', fontsize=28)  # Increased axis label font size
#ax.set_ylabel(r'$p(x)$', fontsize=22)  # Added y-axis label with increased font size
ax.set_title(r'$n = 4$', fontsize=30)  # Increased title font size
#ax.legend(fontsize=16)  # Increased legend font size
ax.grid(False)
ax.set_xlim(x_min, x_max)
# ax.set_ylim(0, max(y_bin_centers_4)*1.2)  # Shared y-limit already set by sharey=True

# Set tick label sizes
ax.tick_params(axis='both', labelsize=24)

# Remove the global title
# plt.suptitle(...)  # Removed as per user request

# Adjust layout to decrease space between subplots
plt.subplots_adjust(wspace=0.35)  # Reduced horizontal space between plots from 0.4 to 0.2

# --- Adding an Arrow Between the Two Subplots ---
# Define the arrow's start and end points in figure coordinates
# (0,0) is bottom-left and (1,1) is top-right of the figure

# Example coordinates: from the center right of the first subplot to the center left of the second subplot
arrow_start = (0.48, 0.5)  # Adjusted if necessary
arrow_end = (0.55, 0.5)    # Reduced the distance between start and end

# Create the FancyArrowPatch
arrow = FancyArrowPatch(arrow_start, arrow_end,
                        arrowstyle='->',
                        mutation_scale=20,  # Size of the arrow
                        color='black',
                        linewidth=3,
                        transform=fig.transFigure)

# Add the arrow to the figure using add_artist
fig.add_artist(arrow)

# Optional: Add a label to the arrow, moved down by adjusting the y-coordinate
fig.text(0.512, 0.56, r'One Iteration', ha='center', va='top', fontsize=22, color='black')  # Moved down from 0.56 to 0.48

plt.savefig(r'C:\Users\0josh\OneDrive\Desktop\Re-organised Honours Year\Thesis Documents\Misc. Plots\Grover_iteration.pdf', dpi=600, bbox_inches='tight',transparent=False)

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad
from matplotlib.patches import FancyArrowPatch

# Enable LaTeX-style text rendering in matplotlib
plt.rcParams['text.usetex'] = True
plt.rcParams['font.family'] = 'serif'

# Define line width and marker size
line_width = 2.5
marker_size = 80  # Note: scatter 's' parameter is in points^2

# 1. Define the Gaussian Function
def gaussian(x, mu=0, sigma=1):
    return (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-((x - mu) ** 2) / (2 * sigma ** 2))

# Parameters
mu = 0              # Mean
sigma = 1           # Standard deviation
x_min, x_max = -3, 3  # Focused range
x = np.linspace(x_min, x_max, 1000)
y = gaussian(x, mu, sigma)

# Function to calculate bin heights as function values at bin centers
def calculate_bin_heights(bin_edges):
    bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2
    y_bin_centers = gaussian(bin_centers, mu, sigma)
    bin_heights = y_bin_centers  # Alternatively, use probability mass
    return bin_centers, bin_heights, y_bin_centers

# Function to calculate bin heights as probability mass within each bin
def calculate_bin_areas(bin_edges):
    bin_areas = np.array([quad(gaussian, bin_edges[i], bin_edges[i+1])[0] for i in range(len(bin_edges)-1)])
    return bin_areas

# 2. Discretize the Gaussian into 8 bins (3-qubit representation) within [-3, 3]
qubits_3 = 3
bins_3 = 2 ** qubits_3  # 8 bins
bin_edges_3 = np.linspace(x_min, x_max, bins_3 + 1)
bin_centers_3, bin_heights_3, y_bin_centers_3 = calculate_bin_heights(bin_edges_3)

# 3. Discretize the Gaussian into 16 bins (4-qubit representation) within [-3, 3]
qubits_4 = 4
bins_4 = 2 ** qubits_4  # 16 bins
bin_edges_4 = np.linspace(x_min, x_max, bins_4 + 1)
bin_centers_4, bin_heights_4, y_bin_centers_4 = calculate_bin_heights(bin_edges_4)

# 4. Plot the Gaussian and both Discretizations Side by Side
fig, axes = plt.subplots(1, 2, figsize=(18, 8), sharey=True)

# Common Gaussian curve for both plots
for ax in axes:
    ax.plot(x, y, label=r'Gaussian', color='#003147', linewidth=line_width)

# --- Left Plot: 3-Qubit (8-Bin) Discretization ---
ax = axes[0]
# Plot discretized bar graph
bar_width_3 = (bin_edges_3[1] - bin_edges_3[0]) * 1  # Width of the bars
ax.bar(bin_centers_3, bin_heights_3, width=bar_width_3,
       alpha=0.6, color='#FED976', edgecolor='black', linewidth=line_width, label=r'$n = 3$')

# Add black dots at each discretization point
ax.scatter(bin_centers_3, y_bin_centers_3, color='black', zorder=5, label='Discretization Points', s=marker_size)

# Labels and Title for Left Plot
ax.set_xlabel(r'$x$', fontsize=34)
ax.set_ylabel(r'$p(x)$', fontsize=34)
ax.set_title(r'$n = 3$', fontsize=34)
# ax.legend(fontsize=16)
ax.grid(False)
ax.set_xlim(x_min, x_max)
ax.set_ylim(0, max(y_bin_centers_4)*1.2)

# Set tick label sizes and x-ticks
ax.tick_params(axis='both', labelsize=34)
ax.set_xticks([-3, -2, -1, 0, 1, 2, 3])

# Remove y-tick values from the first plot
ax.set_yticks([])

# --- Right Plot: 4-Qubit (16-Bin) Discretization ---
ax = axes[1]
# Plot discretized bar graph
bar_width_4 = (bin_edges_4[1] - bin_edges_4[0]) * 1
ax.bar(bin_centers_4, bin_heights_4, width=bar_width_4,
       alpha=0.6, color='#FEB24C', edgecolor='black', linewidth=line_width, label=r'$n = 4$')

# Add black dots at each discretization point
ax.scatter(bin_centers_4, y_bin_centers_4, color='black', zorder=5, label='Discretization Points', s=marker_size)

# Labels and Title for Right Plot
ax.set_xlabel(r'$x$', fontsize=34)
ax.set_ylabel(r'$p(x)$', fontsize=34)  # Added y-axis label
ax.set_title(r'$n = 4$', fontsize=34)
# ax.legend(fontsize=16)
ax.grid(False)
ax.set_xlim(x_min, x_max)

# Set tick label sizes and x-ticks
ax.tick_params(axis='both', labelsize=34)
ax.set_xticks([-3, -2, -1, 0, 1, 2, 3])

# Remove y-tick values from the second plot
ax.set_yticks([])

# Adjust layout to decrease space between subplots
plt.subplots_adjust(wspace=0.35)

# --- Shifting the Arrow and Text to the Left ---
# Define the arrow's start and end points in figure coordinates
arrow_start = (0.47, 0.5)  # Adjusted positions
arrow_end = (0.535, 0.5)

# Create the FancyArrowPatch
arrow = FancyArrowPatch(arrow_start, arrow_end,
                        arrowstyle='->',
                        mutation_scale=20,
                        color='black',
                        linewidth=5,
                        transform=fig.transFigure)

# Add the arrow to the figure using add_artist
fig.add_artist(arrow)

# Shift the text above the arrow
fig.text(0.50, 0.565, r'1 Iter.', ha='center', va='top', fontsize=34, color='black', weight='bold')

plt.savefig(r'C:\Users\0josh\OneDrive\Desktop\Re-organised Honours Year\Thesis Documents\Misc. Plots\Grover_iteration.pdf', dpi=600, bbox_inches='tight', transparent=False)

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import FancyArrowPatch, Rectangle
from scipy.ndimage import gaussian_filter1d
from scipy.interpolate import make_interp_spline

# Configure Matplotlib to use LaTeX-style fonts
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "serif",
    "font.serif": ["Computer Modern"],
    "axes.labelsize": 30,
    "xtick.labelsize": 0,  # Remove x-axis tick labels
    "ytick.labelsize": 0,  # Remove y-axis tick labels
})

# Set parameters
n = 8
num_points = 2**n
x = np.linspace(0, 4 * np.pi, num_points)

# Plot 1: Smooth, Well-Behaved Function (Sine Wave)
y1 = np.sin(x)

# Plot 2: Smooth Random Fluctuations Between 0 and 12
np.random.seed(0)  # For reproducibility
random_noise = np.random.normal(0, 1, num_points)  # Generate random noise

# Apply Gaussian filter to smooth the noise
sigma = 2  # Adjusted for smoother noise
smoothed_noise = gaussian_filter1d(random_noise, sigma=sigma)

# Normalize the smoothed noise to range between 0 and 12
y2_min, y2_max = smoothed_noise.min(), smoothed_noise.max()
y2 = (smoothed_noise - y2_min) / (y2_max - y2_min) * 12

# Plot 3: Smooth, Differentiable Curve Fit to Random Points
np.random.seed(0)  # For reproducibility
y_random = np.random.uniform(-2, 2, num_points)

# Create a cubic spline that interpolates the random points
spline = make_interp_spline(x, y_random, k=3)
y3 = spline(x)

# Create subplots
fig, axs = plt.subplots(1, 3, figsize=(21, 6))

# Plot 1: Smooth Sine Wave
axs[0].plot(x, y1, color='#d1db9c',linewidth=3)
axs[0].set_title(r'\textbf{Highly Structured Vector}', fontsize=30,)  # Added Title
axs[0].set_xlabel(r'$x$',weight='bold')
axs[0].set_ylabel(r'$f(x)$',weight='bold')
axs[0].set_xticks([])  # Remove x-axis ticks
axs[0].set_yticks([])  # Remove y-axis ticks

# Plot 2: Smooth Random Fluctuations Between 0 and 12
axs[1].plot(x, y2, color='#7fcdbb',linewidth=3)
axs[1].set_title(r'\textbf{Transition Period}', fontsize=30)  # Added Title
axs[1].set_xlabel(r'$x$',weight='bold')
axs[1].set_ylabel(r'$f(x)$',weight='bold')
axs[1].set_ylim([0, 12])
axs[1].set_xticks([])  # Remove x-axis ticks
axs[1].set_yticks([])  # Remove y-axis ticks

# Plot 3: Smooth Spline Fit to Random Points
axs[2].plot(x, y3, color='#2c7fb8',linewidth=3)
axs[2].set_title(r'\textbf{Unstructured Vector}', fontsize=30)  # Added Title
axs[2].set_xlabel(r'$x$',weight='bold')
axs[2].set_ylabel(r'$f(x)$',weight='bold')
axs[2].set_xticks([])  # Remove x-axis ticks
axs[2].set_yticks([])  # Remove y-axis ticks

def set_spine_thickness(ax, linewidth=2):
    for spine in ax.spines.values():
        spine.set_linewidth(linewidth)

for ax in axs:
    set_spine_thickness(ax, linewidth=2)

# Adjust layout for better spacing
plt.tight_layout()

# Increase horizontal space to make room for arrows
fig.subplots_adjust(wspace=0.3)

# Function to add an arrow between two axes with padding
def add_arrow(ax1, ax2, fig, text=None, text_pos=(0, 0.05), delta=0.01,shift=-0.01):
    """
    Adds an arrow from ax1 to ax2 on the figure with specified padding.
    """
    bbox1 = ax1.get_position()
    bbox2 = ax2.get_position()

    start_x = bbox1.x1 - delta + shift # Slightly inside the first subplot
    start_y = bbox1.y0 + bbox1.height / 2
    end_x = bbox2.x0 + delta  + shift   # Slightly inside the second subplot
    end_y = bbox2.y0 + bbox2.height / 2

    arrow = FancyArrowPatch(
        (start_x, start_y),
        (end_x, end_y),
        transform=fig.transFigure,
        arrowstyle="->",
        color='black',
        mutation_scale=20,
        linewidth=4
    )
    fig.patches.append(arrow)

    if text:
        fig.text(
            (start_x + end_x) / 2 + text_pos[0]*0.02,
            (start_y + end_y) / 2 + text_pos[1]*0.02,
            text,
            ha='center',
            va='center',
            fontsize=12,
            backgroundcolor='white'
        )

# Add arrows between Plot 1 to Plot 2 and Plot 2 to Plot 3 with padding
add_arrow(axs[0], axs[1], fig, text_pos=(0, 0.05), delta=-0.02)
add_arrow(axs[1], axs[2], fig, text_pos=(0, 0.05), delta=-0.02)

# Function to add a rectangle with text below each plot
def add_rectangle_with_text(ax, label, color, ypos=-0.20):
    """
    Add a colored rectangle below the x-axis of the plot with centered text.
    """
    # Add rectangle (spans 80% of the plot width)
    rect = Rectangle((0.025, ypos), 0.95, 0.1, color=color, transform=ax.transAxes, clip_on=False)
    ax.add_patch(rect)
    # Add text in the center of the rectangle
    ax.text(0.5, ypos + 0.0425, label, ha='center', va='center', fontsize=30, transform=ax.transAxes, color='black', weight='bold')

# Add rectangles below each plot
labels = ['Analytic MPD Algorithm', 'Analytic MPD + Optimisation', 'Inefficient']
colors = ['#66cc66', '#99ccff', '#ff9999']

for ax, label, color in zip(axs, labels, colors):
    add_rectangle_with_text(ax, label, color)

# Function to add arrows between the rectangles below the plots
def add_arrow_between_rectangles(ax1, ax2, fig, ypos=-0.17, delta=0.01,shift=0.0):
    """
    Adds an arrow between the rectangles of ax1 and ax2 below the plot.
    """
    bbox1 = ax1.get_position()
    bbox2 = ax2.get_position()

    # Calculate positions for the arrow (slightly above the center of the rectangles)
    start_x = bbox1.x1 - delta + shift # Right edge of the first rectangle
    start_y = bbox1.y0 + ypos + 0.05  # Just above the rectangle
    end_x = bbox2.x0 + delta + shift  # Left edge of the second rectangle
    end_y = bbox2.y0 + ypos + 0.05  # Just above the rectangle

    # Create the arrow
    arrow = FancyArrowPatch(
        (start_x, start_y),
        (end_x, end_y),
        transform=fig.transFigure,
        arrowstyle="->",
        color='black',
        mutation_scale=20,
        linewidth=4
    )
    fig.patches.append(arrow)

# Add arrows between the rectangles below the plots
add_arrow_between_rectangles(axs[0], axs[1], fig, ypos=-0.175, delta=0.00,shift=0)
add_arrow_between_rectangles(axs[1], axs[2], fig, ypos=-0.175, delta=0.00,shift=0)

plt.savefig(r'C:\Users\0josh\OneDrive\Desktop\Re-organised Honours Year\Thesis Documents\Misc. Plots\encoding_scheme_regimes_with_rectangles_and_arrows.png', dpi=600, bbox_inches='tight', transparent=False)

plt.show()
```
